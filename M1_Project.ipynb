```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Country Wise Capacity of Renewable Energy
df1=pd.read_excel('D:/Module-1(Project)/Data.xlsx','Capacity')
df1.head()
```

```python
# Show years from 2013 to 2023 for world capacity and zoom into 2022 and 2023(for differenciate tham)
years = df1.columns[1:]
fig, axs = plt.subplots(1, 2, figsize=(7, 4))

axs[0].bar(years, df1.loc[0, years]/1000, color='blue',width=0.6)
axs[0].set_title('Globly Year Wise World Capacity')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Capacity(GW)')

axs[1].bar(years, df1.loc[0, years]/1000, color='red',width=0.6)
plt.xlim(2021.5,2023.5)
axs[1].set_title('2022 vs 2023')
axs[1].set_xlabel('Year')

plt.tight_layout()
plt.show()
```

```python
# Read Data notebook from excel file
data=pd.read_excel('D:/Module-1(Project)/Data.xlsx')
data.head()
```

```python
# Take Years from 2001 to 2022
filtereddf = data[(data['Entity']>='World')&(data['Year']>=2001)&(data['Year']<=2022)]
filtereddf.head()
```

```python
df=pd.read_excel('D:/Module-1(Project)/Data.xlsx','Total')
df.head()
```

```python
dff=df[(df['Year']>=2001)&(df['Year']<=2022)]
dff.head()
```

```python
total_e= df[(df['Entity']>='World')&(df['Year']>=2001)&(df['Year']<=2022)]
total_e.head()
```

```python
# Show Globly Year Wise Different Renewable Energies

fig= plt.subplots(figsize=(7, 4))
plt.plot(total_e['Year'],total_e['Total'],color='blue',linewidth=1,label='Total',marker='.')
plt.plot(filtereddf['Year'], filtereddf['Solar generation - TWh'], color='black',label='solar',marker='.')
plt.plot(filtereddf['Year'], filtereddf['Wind generation - TWh'], color='orange',label='wind',marker='.')
plt.plot(filtereddf['Year'], filtereddf['Hydro generation - TWh'], color='green',label='hydro',marker='.')
plt.plot(filtereddf['Year'], filtereddf['Other renewables'], color='red',label='Total',marker='.')

plt.title('Globly Year Wise Renewable Energy')
plt.xlabel('year')
plt.ylabel('Total Renewable Energy')
plt.legend()
plt.grid()
```

```python
# Country wise energy production
selected_countries = ["India", "China", "South Africa", "United States","France", "Japan",'Germany','Africa','Australia','Canada']

df_filtered = dff[dff["Entity"].isin(selected_countries)]

all_years = dff["Year"].unique()

all_years.sort()

plt.figure(figsize=(7, 5))
for country in df_filtered["Entity"].unique():
    df_subset = df_filtered[df_filtered["Entity"] == country]
    plt.plot(all_years, df_subset.set_index("Year")["Total"], label=country, marker='o')  # Use all_years for x-axis

plt.xlabel("Year")
plt.xticks(rotation='vertical')
plt.ylabel("Production(TWh)")
plt.title("Country wise Renewable Energy")
plt.legend()
plt.grid(True)
plt.xticks(all_years)
plt.plot()
```

```python
# Take 2022 year
filtereddf1 = data[(data['Year']==2022)]
countries_of_interest = ['China', 'United States','India','Japan','Germany']
filtered_rows = filtereddf1[filtereddf1['Entity'].isin(countries_of_interest)]
# print(filtered_rows)
```

```python
# Take 2021 year
filtereddf2 = data[(data['Year']==2021)]
countries_of_interest = ['China','United States','India','Japan','Germany']
filtered_rows1 = filtereddf2[filtereddf2['Entity'].isin(countries_of_interest)]
# print(filtered_rows)
```

```python
# Take 2020 year
filtereddf3 = data[(data['Year']==2020)]
countries_of_interest = ['China','United States','India','Japan','Germany']
filtered_rows2 = filtereddf3[filtereddf3['Entity'].isin(countries_of_interest)]
# print(filtered_rows)
```

```python
# Show Countries which consist top positions for Production Of Solar Energy in world(in years of 2020,2021,2022)

fig= plt.subplots(figsize=(7, 5))
plt.bar(np.arange(filtered_rows.shape[0]) -0.2,filtered_rows['Solar generation - TWh'],width=0.2,color='green',label='2022')
plt.bar(np.arange(filtered_rows.shape[0]) ,filtered_rows1['Solar generation - TWh'],width=0.2,color='red',label='2021')
plt.bar(np.arange(filtered_rows1.shape[0]) +0.2,filtered_rows2['Solar generation - TWh'],width=0.2,color='blue',label='2020')
plt.xticks(np.arange(filtered_rows2.shape[0]),filtered_rows['Entity'])
plt.title('Top Solar Energy Producers')
plt.xlabel('Countries')
plt.ylabel('Solar generation - TWh')
plt.legend()
plt.grid()
plt.show()
```
